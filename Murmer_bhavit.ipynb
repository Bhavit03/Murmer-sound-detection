```python
import os
import librosa
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Organize your data
data_dir = r"C:\Users\karnb\Heart Disease\data"
class_folders = os.listdir(data_dir)
class_labels = class_folders  # Assuming class folder names are the labels

# Step 2: Data Preprocessing
def audio_features(filename): 
    sound, sample_rate = librosa.load(filename)
    stft = np.abs(librosa.stft(sound))  

    mfccs = np.mean(librosa.feature.mfcc(y=sound, sr=sample_rate, n_mfcc=40),axis=1)
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate),axis=1)
    mel = np.mean(librosa.feature.melspectrogram(y=sound, sr=sample_rate),axis=1)
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate),axis=1)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(sound), sr=sample_rate),axis=1)

    concat = np.concatenate((mfccs,chroma,mel,contrast,tonnetz))
    return concat

# Step 3: Data Preprocessing Loop
data = []
labels = []

for label in class_labels:
    folder_path = os.path.join(data_dir, label)
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        features = audio_features(file_path)
        data.append(features)
        labels.append(label)

# Step 4: Label Encoding
le = LabelEncoder()
labels = le.fit_transform(labels)

# Step 5: Split Data
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import numpy as np
```

```python
# Normalize the input features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
# Define the neural network model
model = keras.Sequential([
    keras.layers.Input(shape=(193,)),  # Input layer with 4 features
    keras.layers.Dense(128, activation='relu'),  # 1st hidden layer
    keras.layers.Dense(64, activation='relu'),   # 2nd hidden layer
    keras.layers.Dense(32, activation='relu'),   # 3rd hidden layer
    keras.layers.Dense(16, activation='relu'),   # 4th hidden layer
    keras.layers.Dense(5, activation='softmax')  # Output layer with 5 classes
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history=model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test),verbose=1)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
